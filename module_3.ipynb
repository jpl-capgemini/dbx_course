## How to Add a Dashboard to Your Workspace in Databricks

1. In the Databricks workspace, click **Dashboards** in the sidebar.
2. Click **Create Dashboard** or **New Dashboard**.
3. Add visualizations and commentary by selecting queries, charts, or text.
4. Click **Save** to store your dashboard in your workspace.
5. To organize, move the dashboard to a folder in `/Workspace/Users/<your-username>`.

You can access and manage dashboards from the workspace browser, share them with colleagues, and set permissions as needed.

## Task: Build a Multi-Page Dashboard

Create a Databricks dashboard with three pages to explore customer, product, and transaction data. For each page, add the following visualizations:

---

### Page 1: Demographics

- **Gender Breakdown:** Pie or bar chart showing customer gender distribution.
- **Age Group Distribution:** Bar chart of customers by age group.
- **Signup Trend:** Line chart of customer signups over time.
- **Lifetime Value by Age Group:** Bar chart of average customer lifetime value per age group.
- **Customer Segment Distribution:** Pie or bar chart of customer segments.
- **Purchases by Customer Segment:** Bar chart of total purchases per segment.

---

### Page 2: Products

- **Total Units Sold per Product:** Bar chart of units sold for each product.
- **Transaction Quantity Distribution:** Histogram of transaction quantities.

---

### Page 3: Transactions

- **Transaction Distribution per Location:** Map or bar chart of transactions by location.
- **Transaction Distribution per Date:** Line or bar chart of transactions over time.
- **Total Spend per Date:** Line chart of total spend by date.
- **Total Spend per Location:** Bar chart of total spend by location.

---
